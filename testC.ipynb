```python
def myThresh(img,thresh):
    row,cols=img.shape
    for i in range(row):
        for j in range(cols):
            if(img[i,j]>thresh):
                img[i,j]=255.0
            else:
                img[i,j]=0.0
    return img

img=cv2.imread(r'C:\Users\user\Desktop\Advanced-Python-workshop\Advanced-Python-Day-03-main\Samples\gray_image.png',0)
myImg=myThresh(img,80)
plt.imshow(myImg,cmap='gray')
```

```python
frame=cv2.imread('Samples/colors3.jpg')
hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)

lower_color=np.array((96,227,130))
upper_color=np.array((140,255,255))

mask=cv2.inRange(hsv,lower_color,upper_color)   

res=cv2.bitwise_and(frame,frame, mask=mask)

cv2.imshow('FRAME',frame)
cv2.imshow('HSV',hsv)
cv2.imshow('MASK',mask)
cv2.imshow('RES',res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

frame=cv2.imread('Samples/colors3.jpg')

height,width=frame.shape[0:2]
frame=cv2.resize(frame,(int(height/2), int(width/2))

hsv = cv2.cvtColor(frame,cv2.COLOR_BDR2HSV)
                 
lower_color=np.array((109,79,0))
upper_color=np.array((126,255,255))

mask=cv2.inRange(hsv,lower_color,upper_color)

res=cv2.bitwise_and(frame,frame, mask=mask)
                 
cv2.imshow('FRAME', frame)
cv2.imshow('MASK', mask) 
cv2.imshow('RES', res)
cv2.imshow('HSV', hsv)
                 
cv2.waitKey(0)
cv2.destroyAllWindows()
                 
```

```python
import numpy as np

img=cv2.imread('Samples/gray_image.png')
imagray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret,thresh=cv2.threshold(imagray,80,255,0)
contours,hierarchy=cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
print(len(contours))
print(len(contours[6]))
cv2.drawContours(img, contours, 6, (0,255,0), 3)
cv2.imshow('IMG',img)
cv2.imshow('THRESH',thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
